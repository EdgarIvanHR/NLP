```r
library(caret) # matriz de confusion y otras medidas de clasificacion
library(stats) # ajuste de modelos lineales generalizados


endocrino <-read.csv( "/kaggle/input/diabetes/endocrino.csv", header=TRUE)

endocrino<-data.frame(Edad = endocrino$EDAD, Genero = endocrino$SEXO,
                      trigliceridos =  endocrino$TG, creatinina =endocrino$CREATININA,
                      glucosa = endocrino$GLUCB, colesterol = endocrino$COLESTEROL ,diabetes = as.factor(endocrino$DM))
set.seed(9132)
muestra1<-sample(seq(1,length(endocrino[,1])),700)
aux<-endocrino[-muestra1,]
endocrino<-endocrino[muestra1,]
modelo_full<- glm(diabetes ~ Edad + Genero + trigliceridos + creatinina
                  + glucosa + colesterol, data = endocrino, family = "binomial")
summary(modelo_full)

modelo_redu <-glm(diabetes ~ Edad +  glucosa + colesterol, data = endocrino, family = "binomial")

probas_predichas<- predict(modelo_full,aux,type="response" )
etiquetas_predichas<-rep(1,length(probas_predichas))
etiquetas_predichas[which(probas_predichas < .5 )] <- 0
# Matriz de confusion del modelo completo
confusionMatrix(as.factor(etiquetas_predichas),as.factor(aux$diabetes))

modelo_redu1 <-glm(diabetes ~ Edad +  glucosa + colesterol, data = endocrino, family = "binomial")
probas_predichas<- predict(modelo_redu1,aux,type="response" )
etiquetas_predichas<-rep(1,length(probas_predichas))
etiquetas_predichas[which(probas_predichas < .5 )] <- 0
# Matriz de confusion del primer modelo reducido
confusionMatrix(as.factor(etiquetas_predichas),as.factor(aux$diabetes))

modelo_redu2 <-glm(diabetes ~ Edad +  glucosa + colesterol, data = endocrino, family = "binomial")
probas_predichas<- predict(modelo_redu2,aux,type="response" )
etiquetas_predichas<-rep(1,length(probas_predichas))
etiquetas_predichas[which(probas_predichas < .5 )] <- 0
# Matriz de confusion del segundo modelo reducido
confusionMatrix(as.factor(etiquetas_predichas),as.factor(aux$diabetes))
```
